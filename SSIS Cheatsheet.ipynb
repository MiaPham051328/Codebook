# Period Over Period Report

  



``` sql
SELECT YEAR(soh.OrderDate) as 'Order Year'
	 , MONTH (soh.OrderDate) as 'iOrderMonth'
	 , DATENAME (M,soh.OrderDate) as 'Order Month'
	 , FORMAT (soh.OrderDate, 'MMM') as 'Short Month'
	 , SUM (soh.SubTotal) as 'Total Sales'
	 , SUM (soh.TaxAmt) as 'Total Tax'
	 , SUM(soh.Subtotal) + SUM(soh.TaxAmt) as 'Tax In'
FROM stg.SalesOrderHeader soh
WHERE YEAR(soh.OrderDate) BETWEEN @p_year-1 AND  @p_year
GROUP BY YEAR(soh.OrderDate)
	 , MONTH (soh.OrderDate) 
	 , DATENAME (M,soh.OrderDate)
	 , FORMAT (soh.OrderDate, 'MMM')
ORDER BY [Order Year], [iOrderMonth]
```

````sql
DECLARE @p_year int = 2013

SELECT YEAR(soh.OrderDate) as 'Order Year'
	 , MONTH (soh.OrderDate) as 'iOrderMonth'
	 , DATENAME (M,soh.OrderDate) as 'Order Month'
	 , FORMAT (soh.OrderDate, 'MMM') as 'Short Month'
	 , SUM (soh.SubTotal) as 'Total Sales'
	 , SUM (soh.TaxAmt) as 'Total Tax'
	 , SUM(soh.Subtotal) + SUM(soh.TaxAmt) as 'Tax In'
FROM stg.SalesOrderHeader soh
WHERE YEAR(soh.OrderDate) BETWEEN @p_year-1 AND  @p_year
GROUP BY YEAR(soh.OrderDate)
	 , MONTH (soh.OrderDate) 
	 , DATENAME (M,soh.OrderDate)
	 , FORMAT (soh.OrderDate, 'MMM')
ORDER BY [Order Year], [iOrderMonth]
````
